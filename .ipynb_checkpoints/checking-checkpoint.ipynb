```python
############################################ CATHAY ############################################
# read in cathay file
cathay = pd.ExcelFile(FILE_CATHAY)
cathay = cathay.parse(cathay.sheet_names[0])
cathay.head()
print(cathay.dtypes)
```

```python
cathay['date'] = cathay['訂單時間'].apply(lambda x: pd.to_datetime(x[:10], format = "%Y-%m-%d").date())
print(f"Cathay order file has {cathay.shape[0]} rows")
cathay.head()
```

```python
# filter out cashFlow ushop to 國泰世華visa
cashFlow_cathay = cashFlow_USHOP[cashFlow_USHOP["付款方式"] == "國泰世華VISA"]
cashFlow_cathay['cash_id'] = cashFlow_cathay["付款資訊"].apply(lambda x: x[-14:])
cashFlow_cathay.head()
```

```python
print(f"Cashflow with only ushop stores filtered by cathay bank transactions have {cashFlow_cathay.shape[0]} rows")
```

```python
reserve_columns = ['交易平台', '交易序號', '出貨類型', '取消日期', '付款方式', '出貨單號', '交易金額', '配送狀態時間','平台訂單編號', '付款資訊', '建立時間', '訂單編號', '訂單時間', '請/退款金額']
# full merge of two dataframes
cashFlow_cathay_merged = cashFlow_cathay.merge(cathay, left_on = "cash_id", right_on = "訂單編號", how = "outer")
cashFlow_cathay_merged
```

```python
# 'cash_id', '訂單編號'
# createTime, date
# 交易金額, 請/退款金額
# return True if verified, otherwise False
def verifyCathay(df):
    df['verification'] = (df['cash_id'] == df['訂單編號']) & (df['交易金額'] == df['請/退款金額']) & (df['createTime'] == df['date'])
    return df

cashFlow_cathay_merged = verifyCathay(cashFlow_cathay_merged)
cashFlow_cathay_merged
```

```python
cashFlow_cathay_merged_unmatch = cashFlow_cathay_merged[~cashFlow_cathay_merged['verification']]
print(cashFlow_cathay_merged_unmatch.shape)
cashFlow_cathay_record_unmatch = cashFlow_cathay_merged_unmatch[(cashFlow_cathay_merged_unmatch['訂單編號'].notna() & cashFlow_cathay_merged_unmatch['cash_id'].notna())]
cashFlow_cathay_record_rest = cashFlow_cathay_merged_unmatch[~(cashFlow_cathay_merged_unmatch['訂單編號'].notna() & cashFlow_cathay_merged_unmatch['cash_id'].notna())]
```
